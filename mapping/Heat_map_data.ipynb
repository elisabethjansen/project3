```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
# Load the existing CSV file
dirty_plots_df = pd.read_csv('long_lat_test.csv')
dirty_plots_df.(outer)
```

```python
clean_plots_df = dirty_plots_df[['airport', 'longitude', 'latitude']]
clean_plots_df.head()
```

```python
clean_plots_df.to_csv('clean_plots.csv', index=False)
```

```python
import geopandas as gpd
from shapely.geometry import Point

# Convert DataFrame to GeoDataFrame with Point geometry
geometry = [Point(xy) for xy in zip(clean_plots_df['longitude'], clean_plots_df['latitude'])]
gdf = gpd.GeoDataFrame(clean_plots_df, geometry=geometry)
gdf
```

```python
# Save GeoDataFrame to a GeoJSON file
gdf.to_csv('geometry_plots.csv', index=False)
```
