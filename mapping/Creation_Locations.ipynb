```python
import pandas as pd
file_path = r"/Users/poncho/Desktop/heat_map/iata-icao.csv"

messy_df1 = pd.read_csv(file_path)
messy_df1
```

```python
# Filter the DataFrame to only include rows where country_code is 'US'
us_data = messy_df1[messy_df1['country_code'] == 'US']

# Display the filtered data
us_data
```

```python
# Keep only the specified columns in the DataFrame
us_data_filtered = us_data[['country_code', 'iata', 'latitude', 'longitude']]

# Display the filtered data with the selected columns
us_data_filtered
```

```python
# Rename the 'iata' column to 'airport'
us_data_filtered.rename(columns={'iata': 'airport'}, inplace=True)

# Display the DataFrame with the renamed column
us_data_filtered
```

```python
# Drop the 'country_code' column from the DataFrame
us_data_filtered.drop(columns=['country_code'], inplace=True)

# Display the DataFrame after dropping the column
us_data_filtered
```

```python
# Define the file path for the output CSV file
output_file_path = "/Users/poncho/Desktop/heat_map/locations.csv"

# Export the DataFrame to a CSV file named locations.csv without including the index
us_data_filtered.to_csv(output_file_path, index=False)

# Confirm that the export was successful
print("Data exported to locations.csv file successfully.")
```
